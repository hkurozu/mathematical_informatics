```python
import pandas as pd

jobs = pd.DataFrame([
    {"job": "J1", "proc_time": 3, "due": 10},
    {"job": "J2", "proc_time": 6, "due": 12},
    {"job": "J3", "proc_time": 2, "due": 8},
    {"job": "J4", "proc_time": 4, "due": 15},
])

current_time = 5
```

```python
jobs["fifo_rank"] = range(1, len(jobs) + 1)
```

```python
jobs["slack"] = jobs["due"] - (current_time + jobs["proc_time"])
jobs["priority_score"] = jobs["slack"]
```

```python
recommendation = jobs.sort_values("priority_score")

recommendation[["job", "proc_time", "due", "slack"]]
```

```python
def explain(row):
    if row.slack < 0:
        return "すでに遅れリスクあり"
    elif row.slack < 2:
        return "納期が近い"
    else:
        return "比較的余裕あり"

recommendation["reason"] = recommendation.apply(explain, axis=1)
```

```python
recommendation[["job", "slack", "reason"]]
```

```python
def recommend_next_job(jobs, current_time):
    jobs = jobs.copy()
    jobs["slack"] = jobs["due"] - (current_time + jobs["proc_time"])
    jobs = jobs.sort_values("slack")
    return jobs[["job", "proc_time", "due", "slack"]]
```

```python
recommend_next_job(jobs, current_time)
```
